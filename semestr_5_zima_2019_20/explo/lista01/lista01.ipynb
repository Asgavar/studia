```python
def f(sth, a, b):
    return a*sth + b

def dist(x, y, a, b):
    return abs(a*x -y + b) / math.sqrt(a**2 + 1)

points = [
    (1, 10), (2, 10), (3, 11), (4, 12), (5, 18),
    (6, 18), (7, 19), (8, 26), (9, 19), (10, 26),
]
xs, ys = zip(*points)
plt.plot(xs, ys, 'mo')

func_args = np.linspace(1, 10)
func_vals = f(func_args, 2, 5)
plt.plot(func_args, func_vals, 'orange')

distances = [dist(x, y, 2, 5) for x, y in points]
print(f'Starting point (a = 2, b = 5): {sum(distances)}')
min_dist = sum(distances)
min_a = -42
min_b = -42

for a in np.arange(0., 10, 0.01):
    for b in np.arange(3., 10, 0.01):
        curr_dist = sum([dist(x, y, a, b) for x, y in points])
        if curr_dist < min_dist:
            min_dist = curr_dist
            min_a = a
            min_b = b

print(f'Found something better, a = {min_a}, b = {min_b}, dist = {min_dist}')
plt.plot(func_args, min_a * func_args + min_b, 'green')

plt.savefig('zadanie3b.png')
plt.show()
```

```python
def gen_centers(K, d):
    radius = d / (2 * math.sin(180 / K))
    return [
        (radius * math.cos(2 * math.pi * idx / K),
         radius * math.sin(2 * math.pi * idx / K))
        for idx in range(1, K+1)
    ]

def gen_clouds(K, d):
    centers = gen_centers(K, d)
    for center_x, center_y in centers:
        xs = np.random.normal(center_x, 1, 1_000)
        ys = np.random.normal(center_y, 1, 1_000)
        plt.plot(xs, ys, 'k.')
    plt.title(f'{K} groups with side = {d}')
    plt.show()

for K in [7, 11, 23]:
    for d in [5, 10, 15]:
        gen_clouds(K, d)
```
