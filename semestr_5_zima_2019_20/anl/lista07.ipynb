```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def polynomial(n_plus_one, xs):
    return (xs[:, np.newaxis] - np.linspace(-1, 1, n_plus_one - 1)).prod(1)
```

```python
def polynomial_chebyshev(n_plus_one, xs):
    n = n_plus_one - 1
    chebyshev_nodes = np.array([np.cos(((2*k - 1) / (2 * n)) * np.pi) for k in range(1, n+1)])
    return (xs[:, np.newaxis] - chebyshev_nodes).prod(1)
```

```python
xs = np.linspace(-1, 1, 100_00)

for n in range(4, 20+1):
    ys_uniform = polynomial(n+1, xs)
    plt.plot(xs, ys_uniform, color=np.random.rand(3,))
    plt.axhline(y=0, color='orange', linestyle='--')
    plt.title(f'n = {n}, uniform')
    plt.show()
    
    ys_chebyshev = polynomial_chebyshev(n+1, xs)
    plt.plot(xs, ys_chebyshev, color=np.random.rand(3,))
    plt.axhline(y=0, color='orange', linestyle='--')
    plt.title(f'n = {n}, chebyshev')
    plt.show()
```
